```python
sns.set_style("whitegrid")
sns.countplot(y='Category',data=df)
```

```python
table = pd.crosstab(df['Gender'], df['Category'])
sns.heatmap(table, annot=True, fmt='d', cmap='Blues')
```

```python
plt.figure(figsize=(12, 12))
table = pd.crosstab(df['Gender'], df['ItemPurchased'])
sns.heatmap(table, annot=False,cmap='Blues')
```

```python
plt.figure(figsize=(12, 6))
male_df = df[df['Gender'] == 'Male']
table_male = pd.crosstab(male_df['Gender'], male_df['ItemPurchased'])
sns.heatmap(table_male, annot=True, fmt='d', cmap='Blues')
```

```python
plt.figure(figsize=(12, 6))
female_df = df[df['Gender'] == 'Female']
table_female = pd.crosstab(female_df['Gender'], female_df['ItemPurchased'])
sns.heatmap(table_female, annot=True, fmt='d', cmap='Blues')
```
